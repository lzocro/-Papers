## Theory

### The prod algorithm


a)
For $x\in\mathbb{R}^1$ a simple way to prove this inequality is to bound the difference of the two sides by simply differentiating and using continuity arguments. Let $f$ denote the function $x\mapsto \log(x+1)-x+x^2$. $f$ is differentiable on $[-1/2,\infty)$ and its derivative can there be written as the rational fraction $\frac{2x^2+x}{x+1}$.
This is negative on $[-1/2,0]$ and positive on $[0,\infty)$, so if $f(0)\geq 0$ the proof is complete. It is trivially verified that $f(0)=0$, so we have for $x\in[-1/2,\infty)$ $\log(x+1)\geq x-x^2$.

b)
We begin with the logarithm of the weight of a single arm $w_t(k)$:

\[Â \log w_{T+1}(k)= \sum_{t=1}^T\log\left(1+\eta(k)(p_t\cdot l_t-l_t(k))\right)\,.\]
As the losses are bounded in $[0,1]$ and $\eta(k)$ is upper bounded for all $k$ by $1/2$, we can apply the formula from part a) which gives:
\begin{align}
\log w_{T+1}(k)&\geq \sum_{t=1}^T \eta(k)\left(p_t\cdot l_t-l_t(k)\right)-\eta(k)^2\left(p_t\cdot l_t-l_t(k)\right)^2\\
&\geq \eta(k)\sum_{t=1}^T\left(p_t\cdot l_t-l_t(k)\right)-\eta(k)^2\sum_{t=1}^T\left(p_t\cdot l_t-l_t(k)\right)^2\,.
\end{align}

It remains now to translate the result to $W_{T+1}$. See that it is sufficient to show that all $w_t(k)$ are strictly positive for all $t$ and $k$, as then by virtue of the sum being larger than the max and of the logarithm being increasing we have $\log W_{T+1}=\log\sum_{k=1}^Kw_{t+1}(k)\ge \log \max_k{w_{t+1}(k)}$.

This simple by recurrence, as $w_1(k)=1>0$ for all k. Then given the statement is true at $t\ge 1$, we can write for all $k$:
\[ w_{t+1}(k)=w_t\times \left(1+\eta(k)\left(\frac{\eta(k)w_t(k)}{\sum_j \eta(j)w_t(j)}\cdot l_t - l_t(k)\right) \right)\,.\]
We can then lower bound the instant regret component with $-1$, which as $\eta(k)\le 1/2$ guarantees the second term is strictly positive, and thus so is $w_{t+1}(k)$ for all $k$.

Thus for all $k$, $t$, $w_t(k)>0$ and the inequality holds.

c)


As $w_1(k)=1$, it is easily seen that $W_t=K$ for all $t$, so $\log W_{T+1}=\log K$.

d) 

By c) and b), we have:
\[\log K \ge \eta(k)\sum_{t=1}^T\left(p_t\cdot l_t-l_t(k)\right)-\eta(k)^2\sum_{t=1}^T\left(p_t\cdot l_t-l_t(k)\right)^2\,.\]
Which implies:
\[\log K + \eta(k)^2\sum_{t=1}^T\left(p_t\cdot l_t-l_t(k)\right)^2\ge \eta(k)\sum_{t=1}^T\left(p_t\cdot l_t-l_t(k)\right)\,.\]
Dividing by $\eta(k)$ assumed to be non-zero, it remains to optimise $\eta(k)$. For simplicity denote the sum as $y$; we find the maximum of $\frac{\log K}{\eta(k)}+\eta(k)y$ in $\eta_k$, this is simply solved by differentiation and one gets $\eta(k)=\sqrt{\log(K)/y}$. Injection into the previous equation yields two equal terms which we can thus rearrange into:
\[\sum_{t=1}^T\left(p_t\cdot l_t-l_t(k)\right)\le 2\sqrt{\log(K) \sum_{t=1}^T\left(p_t\cdot l_t-l_t(k)\right)^2}\, , \]
as wanted.



### Sleeping Experts


a) Note that by the definition of $\overset{\sim}{l}_t$, we need only show that $\overset{\sim}{p}_t\cdot \overset{\sim}{l}_t = p_t\cdot l_t$ for all $t$. See that for all $t$ we have:
\begin{align}
\overset{\sim}{p}_t\cdot \overset{\sim}{l}_t &= \sum_k \overset{\sim}{p}_t(k) \overset{\sim}{l}_t(k)\\
&=\sum_{k\in A_t} \overset{\sim}{p}_t(k) l_t(k) + \sum_{k\not\in A_t} \overset{\sim}{p}_t(k)\times (p_t\cdot l_t)\\
&=\left(\sum_{k\in A_t}\overset{\sim}{p}_t(k)\right)\times\sum_{k\in A_t} p_t(k)l_t(k) + \left(\sum_{k\not\in A_t}\overset{\sim}{p}_t(k)\right) \times (p_t\cdot l_t)\\
&=\left(\sum_{k\in A_t}\overset{\sim}{p}_t(k)\right)\times (p_t\cdot l_t)+ \left(1 - \sum_{k\in A_t}\overset{\sim}{p}_t(k)\right) \times (p_t\cdot l_t)\\
&= (p_t\cdot l_t)\,.
\end{align}

As wanted.






b)
 See that by 1.1.2.a) we have:
 \begin{align}
 R_T(k)&=\sum_t p_t\cdot l_t - l_t(k)\\
 &=\sum_t \overset{\sim}{p}_t\cdot\overset{\sim}{l}_t - \overset{\sim}{l}_t(k)\\
 &\le 2\sqrt{\log(K) \sum_t\left(\overset{\sim}{p}_t\cdot\overset{\sim}{l}_t - \overset{\sim}{l}_t(k)\right)^2}\label{eq:2}\\
 &=2\sqrt{\log(K) \sum_t\left(p_t\cdot l_t - l_t(k)\right)^2\mathbb{1}\{k\in A_t\}}\label{eq:3}\\
 &\le 2\sqrt{\log(K)\sum_t \mathbb{1}\{k\in A_t\}}\label{eq:4}\\
 &=2\sqrt{\log(K)\sum_t T_k}\,.
 \end{align}
 
 Where \ref{eq:3} uses the property from 1.1.2.a), \ref{eq:2} uses the regret identity from 1.1.1.d) applied to $\overset{\sim}{l}_t$, and \ref{eq:4} uses the boundedness of losses.

```python
#importing data
ideas = np.genfromtxt('ideas_dataset/ideas_votes.csv', delimiter=',', skip_header=1, dtype=int)
politicians= np.genfromtxt('politicians_dataset/politicians_votes.csv', delimiter=',', skip_header=1, dtype=int)
```

```python
#creating a class to handle the learning process
class OO:
    def __init__(self,data, trueloss=False): #one class per dataset
        self.data = data
        self.T = self.data.shape[0]
        self.K = int(2*np.max(self.data[:,1:3])) #2* 39 for pol , 261 for ideas
        self.trueloss = trueloss

    def forecast(self,t):
        #output allowed values (ideas then ideas +K:
        available = [self.data[t,1]-1,self.data[t,2]-1,self.data[t,1]+int(self.K/2)-1,self.data[t,2]+int(self.K/2)-1]
        if self.data[t,1] >= self.data[t,2]:
            predicted = [0,1,0,1] #if z_1 >= z_2 then the strategies are in such an order
        else:    
            predicted = [1,0,1,0]
        return(np.array(available,dtype='int'), np.array(predicted))
    
    def loss(self, y, p, t, available, predicted):

        #loss for active arms
        l = np.array([(1-val)*y + (1-y)*val for val in predicted], dtype='float')
        #take the non zero p only:
        p_comp = p[available]
        #set default filler value (loss of the alg)
        l_tilda = np.ones(self.K)*p_comp.dot(predicted)
        #change the values of the forecast entries to the values of those losses
        l_tilda[available] = l
        
        #compute the loss of the alg itself:
        algloss = (1-p_comp.dot(predicted))*y + (1-y)*p_comp.dot(predicted)
        #compute the true loss
        l_t=None
        if self.trueloss:
            l_t = np.int(((p_comp.dot(predicted)-0.5)*(y-0.5))>0)
        return(l_tilda, algloss, l_t)

    def prod(self, in_eta):
        w = np.ones(self.K)
        p_tilda = np.zeros(self.K)
        p = np.empty(self.K)
        l = np.empty(self.K)
        cum_loss=[]
        alg_l_vec=[]
        alg_l_t=[]
        cumval=0
        eta = np.ones(self.K)*in_eta
        for t in range(self.T):
            
            #Output of the forecasters
            available, predicted = self.forecast(t)
            
            #which arms are active
            vec_available = np.zeros(self.K)
            vec_available[available] = 1
            
            #Computing probs
            p_tilda = eta*w / (eta.dot(w))
            p = vec_available*p_tilda / (vec_available.dot(p_tilda))
            
            yhat=p[available].dot(predicted)
            #get modified loss
            l, alg_l, l_t = self.loss(self.data[t,3], p, t, available, predicted)
            
            #update weights
            w = w.copy()*( np.ones(self.K) + eta*(p.dot(l) - l) ) 
            
            #update various output measures:
            alg_l_vec.append(alg_l)
            cumval+=alg_l
            cum_loss.append(cumval) # is this correct? should maybe output the sleeping instant loss instead by modifying loss?
            if self.trueloss:
                alg_l_t.append(l_t)
        if self.trueloss:
            return(cum_loss,alg_l_vec,alg_l_t)
        else:
            return(cum_loss,alg_l_vec)

    def EWA(self, in_eta):
        w = np.ones(self.K)
        p_tilda = np.zeros(self.K)
        p = np.empty(self.K)
        l= np.empty(self.K)
        cum_loss=[]
        alg_l_vec=[]
        alg_l_t = []
        cumval=0
        eta = np.ones(self.K)*in_eta
        
        for t in range(self.T):
            
            #Output of the forecasters
            available, predicted = self.forecast(t)
            
            #which arms are active
            vec_available = np.zeros(self.K)
            vec_available[available] = 1
            
            #Computing probs
            p_tilda = w / (np.sum(w))
            p = vec_available*p_tilda / (vec_available.dot(p_tilda))
            
            yhat=p[available].dot(predicted)
            #get modified loss
            l, alg_l, l_t = self.loss(self.data[t,3], p, t, available, predicted)
            #update weights
            w = w.copy()*np.exp(-eta*l)
            
            #update various output measures:
            alg_l_vec.append(alg_l)
            cumval+=alg_l
            cum_loss.append(cumval) # is this correct? should maybe output the sleeping instant loss instead by modifying loss?
            if self.trueloss:
                alg_l_t.append(l_t)
        if self.trueloss:
            return(cum_loss,alg_l_vec,alg_l_t)
        else:
            return(cum_loss,alg_l_vec)
        
    def OGD(self, in_eta):
        p_tilda = np.ones(self.K,dtype='float64')/self.K
        p = np.empty(self.K)
        l= np.empty(self.K)
        cum_loss=[]
        alg_l_vec=[]
        alg_l_t = []
        cumval=0
        eta = np.ones(self.K)*in_eta
        
        for t in range(self.T):
            
            #Output of the forecasters
            available, predicted = self.forecast(t)
            
            #which arms are active
            vec_available = np.zeros(self.K)
            vec_available[available] = 1
            
            #Computing probs
            p = vec_available*p_tilda / (vec_available.dot(p_tilda))
            
            yhat=p[available].dot(predicted)
            
            #get modified loss
            l, alg_l, l_t = self.loss(self.data[t,3], p, t, available, predicted)
            
            #update weights
            grad = np.zeros(self.K)
            grad[available] = predicted - 2*predicted*self.data[t,3]
            p_tilda = p_tilda.copy() - eta*grad

            #projection step
            y=np.random.rand(10)
            u = np.sort(y)[::-1]
            #find which ks are valid
            index=[((np.sum(u[:(k+1)])-1)/(k+1))<u[k] for k in range(len(u))]
            K=np.max(np.arange(len(u))[index])+1
            tau=(np.sum(u[:K])-1)/(K)
            proj=[max(i-tau,0) for i in y]        
            
            #update various output measures:
            alg_l_vec.append(alg_l)
            cumval+=alg_l
            #print("cumval", cumval)
            cum_loss.append(cumval) # is this correct? should maybe output the sleeping instant loss instead by modifying loss?
            if self.trueloss:
                alg_l_t.append(l_t)
        if self.trueloss:
            return(cum_loss,alg_l_vec,alg_l_t)
        else:
            return(cum_loss,alg_l_vec)
        
    def Rand(self, in_eta):
        w = np.ones(self.K)
        p_tilda = np.zeros(self.K)
        p = np.empty(self.K)
        l= np.empty(self.K)
        cum_loss=[]
        alg_l_vec=[]
        alg_l_t = []
        cumval=0
        eta = np.ones(self.K)*in_eta
        
        for t in range(self.T):
            
            #Output of the forecasters
            available, predicted = self.forecast(t)
            
            #which arms are active
            vec_available = np.zeros(self.K)
            vec_available[available] = 1
            
            #Computing probs
            w=np.random.rand(self.K)
            p_tilda = w / (np.sum(w))
            p = vec_available*p_tilda / (vec_available.dot(p_tilda))
            
            yhat=p[available].dot(predicted)
            #get modified loss
            l, alg_l, l_t = self.loss(self.data[t,3], p, t, available, predicted)
            #update weights
            
            #update various output measures:
            alg_l_vec.append(alg_l)
            cumval+=alg_l
            cum_loss.append(cumval) # is this correct? should maybe output the sleeping instant loss instead by modifying loss?
        else:
            return(cum_loss,alg_l_vec)
```
